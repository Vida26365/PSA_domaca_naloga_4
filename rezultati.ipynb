# Primerjva časa vstavljanja in časa iskanja v AVL drevesu in v Bently-Saxe transformaciji zgoščene tabele

To je predstavitev rezultatov. Narejena je tudi podrobna analiza časovne zahtevnosti v datoteki [Analiza.md](Analiza.md).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data.csv')
```

Primerjava poteka tako, da najprej vstavimo n naključno urejenih elementov v strukturo. Izmerimo koliko časa traja vstavljanje. Nato v isti strukturi poiščemo m naključnih elementov in izmerimo koliko časa to vzame.                                           
Tako naredimo za obe strukturi in zapišemo podatke v csv.                             

```python

df[["n", "avl_insert_ns",  "bentley_insert_ns"]].plot(x = "n")
plt.legend(["AVL drevo", "Bently-Saxova struktura"])
plt.title("Čas vstavljanja")
plt.xlabel("število elementov")
plt.ylabel("čas [ns]")
plt.show() 
```

Vstavljanje v Bently-Saxovo strukturo za zgoščene tabele vzame precej več časa kot vstavljanje v AVL drevo, saj v AVL drevesu poteka vse le z spreminjanjem nekaj kazalcev, medtem ko je pri zgoščeni tabeli potrebno ustvariti novo zgoščevalno funkcijo, jo evalvirati na vseh elementih, ki so spremenjeni v strukturi (kar je lahko vse od 1 do $ \log(n) $), č ni dovolj dobro poskusimo še enkrat, ... Torej, da se ustvari nova zgoščena tabela, je potrebno veliko več dela, kot le zamenjati $ O(\log(n)) $ kazalcev.

```python
df[["n", "avl_search_ns",  "bentley_search_ns"]].plot(x = "n")
plt.legend(["AVL drevo", "Bently-Saxova struktura"])
plt.title("Čas iskanja")
plt.xlabel("število elementov")
plt.ylabel("čas [ns]")
plt.show()
```

Vidimo, da je Bently-Saxova struktura časovno slabša tako za iskanje, kot tudi za vstavljanje elementov. 

Iskanje v AVL drevesu porabi $ O(\log(n)) $ časa. 

V zgoščenih tabelah je iskanje elementov $ O(1) $, vendar je zaradi Bently-Saxove transformacije, potrebno preiskati $ \log(n) $ tabel, ko želimo najti poljuben element, zato je tudu tu časovna zahtevnost $ O(\log(n)) $ časa. 

Čeprav imata strukturi enako časovno zahtevnost, je iskanje po AVL drevesu hitrejše, saj tam iskaje le sledi kazalcem na otroka node, medtem ko pri Bently-Saxovi transformaciji zgoščene tabele moramo na vsakem nivoju računati dve zgoščevalni funkciji, kar je počasnjeje kot sledenje kazalcem.
